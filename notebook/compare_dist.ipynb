```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
plt.style.use('/Users/sara/PycharmProjects/specparam/notebook/mystyle.mplstyle')
```

```python
# Load the data
my_aN = pd.read_csv('/Users/sara/PycharmProjects/specparam/data/final product/compare/solarlike_astroNN.csv')
my_DM = pd.read_csv('/Users/sara/PycharmProjects/specparam/data/final product/compare/solarlike_DistMass.csv')
my_SH = pd.read_csv('/Users/sara/PycharmProjects/specparam/data/final product/compare/solarlike_StarHorse.csv')
SHxDM = pd.read_csv('/Users/sara/PycharmProjects/specparam/data/final product/compare/SHxDM.csv')
SHxaN = pd.read_csv('/Users/sara/PycharmProjects/specparam/data/final product/compare/SHxaN.csv')
aNxDM = pd.read_csv('/Users/sara/PycharmProjects/specparam/data/final product/compare/aNxDM.csv')
```

```python
fig = plt.figure(figsize=(12, 6.5), dpi=400)
gs = GridSpec(2, 3, figure=fig, hspace=0.35, wspace=0.35)
# 创建颜色映射
cmap = plt.cm.twilight
cmap.set_bad(color='white')  # 设置空值颜色为白色
bins = 30
minnum = 2

lim_min, lim_max = 0, 2

# my x StarHorse
axSH = fig.add_subplot(gs[0, 0])
x_axSH = my_SH['dist50_pc'] / 1000
y_axSH = my_SH['cnn_distance'] / 1000
index = np.where((x_axSH < lim_max) & (y_axSH < lim_max))[0]
x_axSH = x_axSH[index]
y_axSH = y_axSH[index]
r_axSH = x_axSH - y_axSH
r_mean_axSH = np.mean(r_axSH)
r_std_axSH = np.std(r_axSH)
h_axSH, xe_axSH, ye_axSH = np.histogram2d(x_axSH, y_axSH, bins=bins)
# h_axSH = np.ma.masked_where(h_axSH == 0, h_axSH)
h_axSH = np.ma.masked_where(h_axSH <= minnum, h_axSH)

# axSH.scatter(x_axSH, y_axSH, s=3, c='k')
c_axSH = axSH.pcolormesh(xe_axSH, ye_axSH, h_axSH.T, cmap=cmap)
fig.colorbar(c_axSH, ax=axSH)
axSH.text(0.1, 0.78, f'$\mu$ = {r_mean_axSH*1000:.0f}\n$\sigma$ = {r_std_axSH*1000:.0f}', transform=axSH.transAxes)

# my x DistMass
axDM = fig.add_subplot(gs[0, 1])
x_axDM = my_DM['WEIGHTED_DIST'] / 1000
y_axDM = my_DM['cnn_distance'] / 1000
index = np.where((x_axDM < lim_max) & (y_axDM < lim_max))[0]
x_axDM = x_axDM[index]
y_axDM = y_axDM[index]
r_axDM = x_axDM - y_axDM
r_mean_axDM = np.mean(r_axDM)
r_std_axDM = np.std(r_axDM)
h_axDM, xe_axDM, ye_axDM = np.histogram2d(x_axDM, y_axDM, bins=bins)
# h_axDM = np.ma.masked_where(h_axDM == 0, h_axDM)
h_axDM = np.ma.masked_where(h_axDM <= minnum, h_axDM)

# axDM.scatter(x_axDM, y_axDM, s=3, c='k')
c_axDM = axDM.pcolormesh(xe_axDM, ye_axDM, h_axDM.T, cmap=cmap)
fig.colorbar(c_axDM, ax=axDM)
axDM.text(0.1, 0.78, f'$\mu$ = {r_mean_axDM*1000:.0f}\n$\sigma$ = {r_std_axDM*1000:.0f}', transform = axDM.transAxes)

# my x astroNN
axaN = fig.add_subplot(gs[0, 2])
x_axaN = my_aN['weighted_dist'] / 1000
y_axaN = my_aN['cnn_distance'] / 1000
index = np.where((x_axaN < lim_max) & (y_axaN < lim_max))[0]
x_axaN = x_axaN[index]
y_axaN = y_axaN[index]
r_axaN = x_axaN - y_axaN
r_mean_axaN = np.mean(r_axaN)
r_std_axaN = np.std(r_axaN)
h_axaN, xe_axaN, ye_axaN = np.histogram2d(x_axaN, y_axaN, bins=bins)
# h_axaN = np.ma.masked_where(h_axaN == 0, h_axaN)
h_axaN = np.ma.masked_where(h_axaN <= minnum, h_axaN)

# axaN.scatter(x_axaN, y_axaN, s=3, c='k')
c_axaN = axaN.pcolormesh(xe_axaN, ye_axaN, h_axaN.T, cmap=cmap)
fig.colorbar(c_axaN, ax=axaN)
axaN.text(0.1, 0.78, f'$\mu$ = {r_mean_axaN*1000:.0f}\n$\sigma$ = {r_std_axaN*1000:.0f}', transform = axaN.transAxes)

# StarHorse x DistMass
axSHDM = fig.add_subplot(gs[1, 0])
x_axSHDM = SHxDM['dist50_pc'] / 1000
y_axSHDM = SHxDM['WEIGHTED_DIST'] / 1000
index = np.where((x_axSHDM < lim_max) & (y_axSHDM < lim_max))[0]
x_axSHDM = x_axSHDM[index]
y_axSHDM = y_axSHDM[index]
r_axSHDM = x_axSHDM - y_axSHDM
r_mean_axSHDM = np.mean(r_axSHDM)
r_std_axSHDM = np.std(r_axSHDM)
h_axSHDM, xe_axSHDM, ye_axSHDM = np.histogram2d(x_axSHDM, y_axSHDM, bins=bins)
# h_axSHDM = np.ma.masked_where(h_axSHDM == 0, h_axSHDM)
h_axSHDM = np.ma.masked_where(h_axSHDM <= minnum, h_axSHDM)

# axSHDM.scatter(x_axSHDM, y_axSHDM, s=3, c='k')
c_axSHDM = axSHDM.pcolormesh(xe_axSHDM, ye_axSHDM, h_axSHDM.T, cmap=cmap)
fig.colorbar(c_axSHDM, ax=axSHDM)
axSHDM.text(0.1, 0.78, f'$\mu$ = {r_mean_axSHDM*1000:.0f}\n$\sigma$ = {r_std_axSHDM*1000:.0f}', transform = axSHDM.transAxes)

# StarHorse x astroNN
axSHaN = fig.add_subplot(gs[1, 1])
x_axSHaN = SHxaN['dist50_pc'] / 1000
y_axSHaN = SHxaN['weighted_dist'] / 1000
index = np.where((x_axSHaN < lim_max) & (y_axSHaN < lim_max))[0]
x_axSHaN = x_axSHaN[index]
y_axSHaN = y_axSHaN[index]
r_axSHaN = x_axSHaN - y_axSHaN
r_mean_axSHaN = np.mean(r_axSHaN)
r_std_axSHaN = np.std(r_axSHaN)
h_axSHaN, xe_axSHaN, ye_axSHaN = np.histogram2d(x_axSHaN, y_axSHaN, bins=bins)
# h_axSHaN = np.ma.masked_where(h_axSHaN == 0, h_axSHaN)
h_axSHaN = np.ma.masked_where(h_axSHaN <= minnum, h_axSHaN)

# axSHaN.scatter(x_axSHaN, y_axSHaN, s=3, c='k')
c_axSHaN = axSHaN.pcolormesh(xe_axSHaN, ye_axSHaN, h_axSHaN.T, cmap=cmap)
fig.colorbar(c_axSHaN, ax=axSHaN)
axSHaN.text(0.1, 0.78, f'$\mu$ = {r_mean_axSHaN*1000:.0f}\n$\sigma$ = {r_std_axSHaN*1000:.0f}', transform = axSHaN.transAxes)

# astroNN x DistMass
axaNDM = fig.add_subplot(gs[1, 2])
x_axaNDM = aNxDM['weighted_dist_aN'] / 1000
y_axaNDM = aNxDM['WEIGHTED_DIST_DM'] / 1000
index = np.where((x_axaNDM < lim_max) & (y_axaNDM < lim_max))[0]
x_axaNDM = x_axaNDM[index]
y_axaNDM = y_axaNDM[index]
r_axaNDM = x_axaNDM - y_axaNDM
r_mean_axaNDM = np.mean(r_axaNDM)
r_std_axaNDM = np.std(r_axaNDM)
h_axaNDM, xe_axaNDM, ye_axaNDM = np.histogram2d(x_axaNDM, y_axaNDM, bins=bins)
# h_axaNDM = np.ma.masked_where(h_axaNDM == 0, h_axaNDM)
h_axaNDM = np.ma.masked_where(h_axaNDM <= minnum, h_axaNDM)

# axaNDM.scatter(x_axaNDM, y_axaNDM, s=3, c='k')
c_axaNDM = axaNDM.pcolormesh(xe_axaNDM, ye_axaNDM, h_axaNDM.T, cmap=cmap)
fig.colorbar(c_axaNDM, ax=axaNDM)
axaNDM.text(0.1, 0.78, f'$\mu$ = {r_mean_axaNDM*1000:.0f}\n$\sigma$ = {r_std_axaNDM*1000:.0f}', transform = axaNDM.transAxes)


axSH.set_xlim(lim_min, lim_max)
axSH.set_ylim(lim_min, lim_max)
axDM.set_xlim(lim_min, lim_max)
axDM.set_ylim(lim_min, lim_max)
axaN.set_xlim(lim_min, lim_max)
axaN.set_ylim(lim_min, lim_max)

axSHDM.set_xlim(lim_min, lim_max)
axSHDM.set_ylim(lim_min, lim_max)
axSHaN.set_xlim(lim_min, lim_max)
axSHaN.set_ylim(lim_min, lim_max)
axaNDM.set_xlim(lim_min, lim_max)
axaNDM.set_ylim(lim_min, lim_max)

axSH.set_xlabel('StarHorse [kpc]')
axDM.set_xlabel('DistMass [kpc]')
axaN.set_xlabel('astroNN [kpc]')
axSHDM.set_xlabel('StarHorse [kpc]')
axSHaN.set_xlabel('StarHorse [kpc]')
axaNDM.set_xlabel('astroNN [kpc]')

axSH.set_ylabel('This work')
axDM.set_ylabel('This work')
axaN.set_ylabel('This work')
axSHDM.set_ylabel('DistMass')
axSHaN.set_ylabel('astroNN')
axaNDM.set_ylabel('DistMass')

axSH.plot([lim_min, lim_max], [lim_min, lim_max], color='k', linestyle=':', linewidth=1.5)
axDM.plot([lim_min, lim_max], [lim_min, lim_max], color='k', linestyle=':', linewidth=1.5)
axaN.plot([lim_min, lim_max], [lim_min, lim_max], color='k', linestyle=':', linewidth=1.5)
axSHDM.plot([lim_min, lim_max], [lim_min, lim_max], color='k', linestyle=':', linewidth=1.5)
axSHaN.plot([lim_min, lim_max], [lim_min, lim_max], color='k', linestyle=':', linewidth=1.5)
axaNDM.plot([lim_min, lim_max], [lim_min, lim_max], color='k', linestyle=':', linewidth=1.5)

axSH.plot([lim_min, lim_max], [lim_min + r_std_axSH, lim_max + r_std_axSH], color='grey', linestyle='--', linewidth=1.5)
axSH.plot([lim_min, lim_max], [lim_min - r_std_axSH, lim_max - r_std_axSH], color='grey', linestyle='--', linewidth=1.5)
axDM.plot([lim_min, lim_max], [lim_min + r_std_axDM, lim_max + r_std_axDM], color='grey', linestyle='--', linewidth=1.5)
axDM.plot([lim_min, lim_max], [lim_min - r_std_axDM, lim_max - r_std_axDM], color='grey', linestyle='--', linewidth=1.5)
axaN.plot([lim_min, lim_max], [lim_min + r_std_axaN, lim_max + r_std_axaN], color='grey', linestyle='--', linewidth=1.5)
axaN.plot([lim_min, lim_max], [lim_min - r_std_axaN, lim_max - r_std_axaN], color='grey', linestyle='--', linewidth=1.5)
axSHDM.plot([lim_min, lim_max], [lim_min + r_std_axSHDM, lim_max + r_std_axSHDM], color='grey', linestyle='--', linewidth=1.5)
axSHDM.plot([lim_min, lim_max], [lim_min - r_std_axSHDM, lim_max - r_std_axSHDM], color='grey', linestyle='--', linewidth=1.5)
axSHaN.plot([lim_min, lim_max], [lim_min + r_std_axSHaN, lim_max + r_std_axSHaN], color='grey', linestyle='--', linewidth=1.5)
axSHaN.plot([lim_min, lim_max], [lim_min - r_std_axSHaN, lim_max - r_std_axSHaN], color='grey', linestyle='--', linewidth=1.5)
axaNDM.plot([lim_min, lim_max], [lim_min + r_std_axaNDM, lim_max + r_std_axaNDM], color='grey', linestyle='--', linewidth=1.5)
axaNDM.plot([lim_min, lim_max], [lim_min - r_std_axaNDM, lim_max - r_std_axaNDM], color='grey', linestyle='--', linewidth=1.5)

plt.savefig('../figure/comp_dist.pdf', bbox_inches='tight')
```
